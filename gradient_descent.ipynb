```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns 
sns.set()
```

```python
x = np.arange(-20, 20, 0.1)
loss = x**2 + 2*x - 4
```

```python
point = -14
plt.figure(figsize=(12, 10))
plt.plot(point, point ** 2 + 2 * point - 4, 'ro')
plt.plot(x, loss)
```

```python
def gd(df=lambda x: 2 * x + 2, learning_rate=0.01, max_iters=1000, precision=0.00001, x_0=-1):
    iters = 1
    prev_step = 1
    points = []
    while prev_step > precision and iters < max_iters:
        x_prev = x_0
        x_0 = x_0 - learning_rate * df(x_prev)
        prev_step = abs(x_0 - x_prev)
        iters += 1
        points.append(x_0)
        print('Iteracja: #{}, Pozycja: {}'.format(iters, x_0))
    print('Minimum: {}'.format(x_0))
    return points
```

```python
points = gd(x_0 = -14, learning_rate=0.1)
```

```python
point = -14
plt.figure(figsize=(12, 10))
plt.plot(x, loss)
for point in points:
    plt.plot(point, point ** 2 + 2 * point - 4, 'ro')
```
