```python
columns2 = ['Transmission Type', 'Driven_Wheels', 'Vehicle Size', 'Vehicle Style']

from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()
ds1[columns2] = encoder.fit_transform(ds1[columns2])
```

```python
X = ds1.drop(columns=['MSRP'], axis=1)
y = ds1['MSRP']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 2)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
```

```python
lin_reg.fit(X_train, y_train)
```

```python
training_data_pred = lin_reg.predict(X_train)
```

```python
from sklearn.metrics import r2_score
error_score = r2_score(y_train, training_data_pred)
print(error_score)
```

```python
plt.figure(figsize=(20, 20))
plt.scatter(y_train, training_data_pred)
plt.show()
```
